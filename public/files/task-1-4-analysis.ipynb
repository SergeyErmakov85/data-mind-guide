```python
data.dropna(subset=['СРЕДНИЙ_БАЛЛ'], inplace=True)
print(data[['ИНТЕЛЛЕКТ', 'МОТИВАЦИЯ', 'СРЕДНИЙ_БАЛЛ', 'КЛАСС']].info())
```

## Identify unique classes and generate pairs

### Subtask:
Extract unique values from the 'КЛАСС' column and create all possible unique pairs of these classes for comparison.


**Reasoning**:
To extract unique values from the 'КЛАСС' column and generate all possible unique pairs of these classes, I will first get the unique values from the 'КЛАСС' column of the `data` DataFrame. Then, I will import the `itertools` module, which contains the `combinations` function needed to generate all unique pairs of these classes.



```python
import itertools

unique_classes = data['КЛАСС'].unique()
class_pairs = list(itertools.combinations(unique_classes, 2))

print(f"Unique Classes: {unique_classes}")
print(f"Class Pairs: {class_pairs}")
```

## Perform Mann-Whitney U test for each pair and variable

### Subtask:
Iterate through each pair of classes and for each specified variable ('ИНТЕЛЛЕКТ', 'МОТИВАЦИЯ', 'СРЕДНИЙ_БАЛЛ'), perform the Mann-Whitney U test. Collect the U statistic and p-value from each test.


**Reasoning**:
I will now iterate through each unique pair of classes and for each specified variable, perform the Mann-Whitney U test, collecting the U statistic and p-value as requested by the subtask instructions.



```python
variables_to_test = ['ИНТЕЛЛЕКТ', 'МОТИВАЦИЯ', 'СРЕДНИЙ_БАЛЛ']
mw_results = []

for class1, class2 in class_pairs:
    class1_data = data[data['КЛАСС'] == class1]
    class2_data = data[data['КЛАСС'] == class2]

    for variable in variables_to_test:
        # Perform Mann-Whitney U test
        u_statistic, p_value = mannwhitneyu(class1_data[variable], class2_data[variable])

        # Store results
        mw_results.append({
            'Class 1': class1,
            'Class 2': class2,
            'Variable': variable,
            'U-statistic': u_statistic,
            'p-value': p_value
        })

print(f"Collected {len(mw_results)} test results.")
```

```python
import pandas as pd

mw_results_df = pd.DataFrame(mw_results)
display(mw_results_df)
```

```python
significant_results_df = mw_results_df[mw_results_df['p-value'] < 0.05]
display(significant_results_df)
```
